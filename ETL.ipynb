# Características
* adult : Indica si la película tiene califiación X, exclusiva para adultos.
* belongs_to_collection : Un diccionario que indica a que franquicia o serie de películas pertenece la película
* budget : El presupuesto de la película, en dólares
* genres : Un diccionario que indica todos los géneros asociados a la película
* homepage : La página web oficial de la película
* id : ID de la película
* imdb_id : IMDB ID de la película
* original_language : Idioma original en la que se grabo la película
* original_title : Titulo original de la película
* overview : Pequeño resumen de la película
* popularity : Puntaje de popularidad de la película, asignado por TMDB (TheMoviesDataBase)
* poster_path : URL del póster de la película
* production_companies : Lista con las compañias productoras asociadas a la película
* production_countries : Lista con los países donde se produjo la película
* release_date : Fecha de estreno de la película
* revenue : Recaudación de la película, en dolares
* runtime : Duración de la película, en minutos
* spoken_languages : Lista con los idiomas que se hablan en la película
* status : Estado de la película actual (si fue anunciada, si ya se estreno, etc)
* tagline : Frase celebre asociadaa la película
* title : Titulo de la película
* video : Indica si hay o no un trailer en video disponible en TMDB
* vote_average : Puntaje promedio de reseñas de la película
* vote_count : Números de votos recibidos por la película, en TMDB


# Transformación de los Datos * Desanidar los Datos

```python
# DESANIDAMOS LA COLUMNA belongs_to_collection
movies_dataset['belongs_to_collection'] = movies_dataset['belongs_to_collection'].fillna('{}') # SUSTITUIMOS fillna PARA RELLENAR LOS VALORES FALTANTES POR UN DICCIONARIO VACIO '{}'
movies_dataset.reset_index(drop=True) # RESTAURAMOS LOS INDICES
dataf1 = pd.json_normalize(movies_dataset['belongs_to_collection'].map(eval)) # ORGANIZAMOS LOS DATOS
dataf1 = dataf1.apply(pd.Series) # TRANSFORMAMOS CADA FILA EN UNA COLUMNA PARA CAMBIAR EL dataf1 EN UN DATAFRAME
dataf1
```

```python
# RENOMBRAMOS LAS COLUMNAS id / name
# MODIFICAMOS EL DATAFRAME ORIGINAL CON EL PARAMETRO inplace=True
#  ELIMINAMOS LAS COLUMNAS CON EL METODO drop
dataf1.rename(columns={'id':'id_collection', 'name':'name_collection'}, inplace=True)
dataf1.drop(columns=['poster_path','backdrop_path'], inplace = True)
```

```python
# REMPLAZAMOS LOS VALORES VACIOS
dataf1.replace('', np.nan, inplace=True)
```

```python
# LLAMAMOS EL DATAFRAME CREADO
dataf1
```

```python
# DESANIDAMOS LA COLUMNA genres
movies_dataset['genres'] = movies_dataset['genres'].fillna('{}') # SUSTITUIMOS fillna PARA RELLENAR LOS VALORES FALTANTES POR UN DICCIONARIO VACIO '{}'
dataf2  = pd.json_normalize(movies_dataset['genres'].map(eval)) # NORMALIZAMOS LOS DATOS CON json Y LO CONVERTIMOS EN UN DATAFRAME ESTRUCTURADO
dataf2
```

```python
# ACCEDEMOS AL VALOR UBICADO EN LA FILA 0 Y LA COLUMNA 1 
dataf2[1][0] 
```

```python
# CREE ESTE CODIGO PARA AGREGAR NUEVAS COLUMNAS AL dataf2 A PARTIR DE LOS VALORES EXTRAIDOS DE LAS COLUMNAS id y name, DE CADA ELEMENTO DEL DATAFRAME dataf2 EN UN BUCLE
Id = []
Name = []
for i in range(8):
    nid = 'id' + str(i)
    nam = 'name' + str(i)
    Id.append(nid)
    Name.append(nam)
    dataf2[nid] = pd.json_normalize(dataf2[i])['id']
    dataf2[nam] = pd.json_normalize(dataf2[i])['name']
```

```python
# ELIMINAMOS LAS COLUMNAS DE LA LISTA col Y LAS REMPLAZAMOS POR VALORES VACIOS
col = list(range(8))
dataf2.drop(columns=col, inplace=True)
dataf2.replace('', np.nan, inplace=True)
```

```python
# CREE 2 NUEVAS COLUMNAS CON VALORES CONCATENADOS
dataf2['id_genres']   = dataf2[Id].apply( lambda x: ',' .join(x.dropna().astype(str)),axis=1)
dataf2['name_genres'] = dataf2[Name].apply(lambda x: ',' .join(x.dropna().astype(str)),axis=1)
```

```python
# ELIMINO LAS COLUMNAS id, Name, DIRECTAMENTE DEL DATAFRAME
dataf2.drop(columns= Id, inplace=True)
dataf2.drop(columns= Name, inplace=True)
```

```python
# LLAMAMOS EL DATAFRAME CREADO
dataf2
```

```python
# REMPLAZAMOS LOS VALORES VACIOS
dataf2.replace('', np.nan, inplace=True)
```

```python
# DESANIDAMOS LA COLUMNA production_companies
movies_dataset['production_companies'] = movies_dataset['production_companies'].fillna('{}') # SUSTITUIMOS fillna PARA RELLENAR LOS VALORES FALTANTES POR UN DICCIONARIO VACIO '{}'
dataf3 = pd.json_normalize(movies_dataset['production_companies'].map(eval)) # NORMALIZAMOS LOS DATOS CON json Y LO CONVERTIMOS EN UN DATAFRAME ESTRUCTURADO
dataf3
```

```python
col = list(range(26))
dataf3.drop(columns=col, inplace=True)
```

```python
dataf3['name_production'] = dataf3[Name].apply(lambda x: ',' .join(x.dropna().astype(str)), axis=1)
dataf3['id_production'] =   dataf3[Id].apply(lambda x: ','   .join(x.dropna().astype(str)), axis=1) 
```

```python
dataf3.drop(columns=Name, inplace=True)
dataf3.drop(columns=Id,   inplace=True)
```

```python
dataf3.replace('', np.nan, inplace=True)
```

```python
dataf3
```

```python
# DESANIDAMOS LA COLUMNA production_countries
movies_dataset['production_countries'] = movies_dataset['production_countries'].fillna('{}')
dataf4 = pd.json_normalize(movies_dataset['production_countries'].map(eval))
dataf4

```

```python
Iso = []
Name = []
for i in range(25):
    niso = 'iso_3166_1' + str(i)
    nam = 'name' + str(i)
    Iso.append(niso)
    Name.append(nam)
    dataf4[niso] = pd.json_normalize(dataf4[i])['iso_3166_1']
    dataf4[nam] = pd.json_normalize(dataf4[i])['name']
```

```python
dataf4['iso_countrie'] =  dataf4[Iso].apply (lambda x: ','   .join(x.dropna().astype(str)),axis=1)
dataf4['name_countrie'] = dataf4[Name].apply(lambda x: ','   .join(x.dropna().astype(str)),axis=1)
```

```python
dataf4.drop( columns=Name, inplace=True)
dataf4.drop( columns=Iso,  inplace=True)
dataf4
```

```python
dataf4.replace('', np.nan, inplace=True)
```

```python
# DESANIDAMOS LA COLUMNA spoken_languages
movies_dataset['spoken_languages'] = movies_dataset['spoken_languages'].fillna('{}')
dataf5 = pd.json_normalize(movies_dataset['spoken_languages'].map(eval))
dataf5
```

```python
Iso = []
Name = []
for i in range(19):
    iso = 'iso_639_1' + str(i)
    nam = 'name' + str(i)
    Iso.append(iso)
    Name.append(nam)
    dataf5[iso] = pd.json_normalize(dataf5[i])['iso_639_1']
    dataf5[nam] = pd. json_normalize(dataf5[i])['name']
```

```python
col = list(range(19))
dataf5.drop(columns=col, inplace = True)
```

```python
dataf5['iso_languague'] = dataf5[Iso].apply(lambda x: ',' .join(x.dropna().astype(str)), axis=1)
dataf5['name_languague'] = dataf5[Name].apply(lambda x: ',' .join(x.dropna().astype(str)), axis=1)
```

```python
movies_dataset_concat = pd.concat([movies_dataset,dataf1,dataf2,dataf3,dataf4,dataf5], axis=1)
```

```python
movies_dataset_concat = movies_dataset_concat.drop(columns=['belongs_to_collection','spoken_languages', 'production_countries', 'genres', 'production_companies'])
```

```python
# RELLENAR LOS VALORES NULOS POR EL NUMERO 0 DE LOS CAMPOS revenue Y budget
movies_dataset_concat['revenue'] = movies_dataset_concat['revenue'].replace(np.nan, 0)
movies_dataset_concat['budget'] = movies_dataset_concat['budget'].replace(np.nan, 0)
```

```python
# ELIMINAR VALORES NULOS DEL CAMPO release_date
movies_dataset_concat['release_date'] = movies_dataset_concat['release_date'].dropna()
```

```python
# FORMATO FECHA AAAA-mm-dd CREAR COLUMNA release_year EXTRAER EL AÑO DE LA FECHA DE ESTRENO 
movies_dataset_concat['release_date'] = pd.to_datetime(movies_dataset_concat['release_date'], format= '%Y-%m-%d', errors='coerce')
movies_dataset_concat['release_year'] = movies_dataset_concat['release_date'].dt.year
```

```python
# COLUMNA RETORNO DE INVERSION LLAMADA return CON LOS CAMPOS revenue Y budget / DIVIDIR ESTAS DOS ULTIMAS COLUMNAS CUANDO NO HAY DATOS DISPONIBLES PARA CALCULARLO, DEBE TOMAR EL VALOR 0
movies_dataset_concat['budget'] = pd.to_numeric(movies_dataset_concat['budget'], downcast = 'float', errors = 'coerce')
movies_dataset_concat['return'] = (movies_dataset_concat['revenue']).divide(movies_dataset_concat['budget']).fillna(0)
movies_dataset_concat['return'].replace([np.inf, -np.inf], np.nan, inplace=True)
movies_dataset_concat['return'].fillna(0, inplace=True)
```

```python
# ELIMINAR LAS COLUMNAS QUE NO SERAN UTILIZADAS
movies_dataset_concat = movies_dataset_concat.drop(columns=['video', 'imdb_id', 'adult', 'original_title', 'poster_path', 'homepage'])
```

## TRANSFORMAMOS EL SEGUNDO DATASETS credits

```python
credits.head()
```

```python
# DESANIDAMOS LA COLUMNA cast
credits['cast'] = credits['cast'].fillna('{}')
Dataframe1 = pd.json_normalize(credits['cast'].map(eval))
Dataframe1
```

```python
# ACCEDEMOS AL VALOR UBICADO EN LA FILA 1 Y LA COLUMNA 0 
Dataframe1[0][1]
```

```python
# AGREGAR NUEVAS COLUMNAS AL DATAFRAME A PARTIR DE LOS VALORES EXTRAIDOS 'character', 'name' y 'gender' DE CADA ELEMENTO DEL dataframe1 EN UN BUCLE 
Character = []
Name = []
Gender = []
for i in range(313):
    char = 'character' + str(i)
    nam = 'name' + str(i)
    gen = 'gender' + str(i)
    Character.append(char)
    Name.append(nam)
    Gender.append(gen)
    Dataframe1[char] = pd.json_normalize(Dataframe1[i])['character']
    Dataframe1[nam]  = pd.json_normalize(Dataframe1[i])['name']
    Dataframe1[gen]  = pd.json_normalize(Dataframe1[i])['gender']
```

```python
# ELIMINAMOS LAS COLUMNAS IDENTIFICADAS DE LA LISTA col
col = list(range(313))
Dataframe1.drop(columns=col, inplace=True)
```

```python
Dataframe1['character']  =   Dataframe1[Character].apply(lambda x: ',' .join(x.dropna().astype(str)), axis= 1)
Dataframe1['name_actor'] =   Dataframe1[Name]     .apply(lambda x: ',' .join(x.dropna().astype(str)), axis= 1)
Dataframe1['genres_actor'] = Dataframe1[Gender]   .apply(lambda x: ',' .join(x.dropna().astype(str)), axis= 1) 
```

```python
Dataframe1.drop(columns=Character, inplace=True)
Dataframe1.drop(columns=Name, inplace=True)
Dataframe1.drop(columns=Gender, inplace=True)
Dataframe1
```

```python
# DESANIDAMOS LA COLUMNA crew
credits['crew'][0]
credits['crew'] = credits['crew'].fillna('{}')
Dataframe2 = pd.json_normalize(credits['crew'].map(eval))
Dataframe2
```

```python
Departament = []
Job = []
Name = []
for i in range(435):
    dep = 'department' + str(i)
    job = 'job' + str(i)
    nam = 'name' + str(i)
    Departament.append(dep)
    Job.append(job)
    Name.append(nam)
    Dataframe2[dep] = pd.json_normalize(Dataframe2[i])['department']
    Dataframe2[job] = pd.json_normalize(Dataframe2[i])['job']
    Dataframe2[nam] = pd.json_normalize(Dataframe2[i])['name']
```

```python
col = list(range(434))
Dataframe2.drop(columns=col, inplace=True)
```

```python
Dataframe2['departament'] = Dataframe2[Departament].apply(lambda x: ',' .join(x.dropna().astype(str)),axis=1)
Dataframe2['job'] = Dataframe2[Job].apply(lambda x: '' .join(x.dropna().astype(str)),axis=1)
Dataframe2['name_director'] = Dataframe2[Name].apply(lambda x: ',' .join(x.dropna().astype(str)), axis=1)
```

```python
Dataframe2.drop(columns=Departament, inplace=True)
Dataframe2.drop(columns=Job, inplace=True)
Dataframe2.drop(columns=Name, inplace=True)
```
